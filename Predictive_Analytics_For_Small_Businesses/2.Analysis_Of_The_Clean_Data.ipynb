```python
# DATA ANALYSIS AND VISUALIZATION
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Load the data from the cleaned Data CSV file
data = pd.read_csv("Cleaned_Data.csv")
```

```python
# Set the size of the graph
plt.figure(figsize=(16,6))
# Create a lineplot 
sns.lineplot(x="date", y="price", hue="commodity", data=data)
# Add a title to the graph
plt.title("National Price Trends Between 2006 and 2026")
plt.show()
```

```python
# Find the top 10 commodities by frequency in the dataset(most data points)
top_commodities =  data["commodity"].value_counts().nlargest(10).index
# data["commodity"].value_counts().nlargest(10).index counts how many times each food item appears in the dataset and selects the top 10.
# The . index saves just the names into the variable top_commodities
# Filter the dataset to include only the top 10 commodities
top_data = data[data["commodity"].isin(top_commodities)]
#.isin(top_commodities) creates a boolean mask that is True for rows where the commodity is in the top_commodities list, and False otherwise
# Create a line plot to show the price trends of the top 10 commodities in the dataset
plt.figure(figsize=(16,6))
sns.lineplot(x="date", y="price", hue="commodity", data=top_data)
# Add a title to the graph
plt.title("Price Trends of the top 10 Commodities")
# Display the line graph
plt.show()
```

```python

# The county comparison graph using seaborn's barplot
# Set the size of the graph
plt.figure(figsize=(16,6))
# Create a bar plot
sns.barplot(x="County", y="price", hue="commodity", data=data)
# Add a title to the graph
plt.title("Price Comparison by County and Commodity")
# Display the graph
plt.show()
```

```python
# Calculate the avaerage price for each county
average_price_by_county = data.groupby("County")["Price"].mean().sort_values(ascending=False).reset_index()
# .groupby("County") groups the data by the "County" column, and ["price"] selects the price of each commodity
# .mean() calculates the average prices for each county
# .sort_values(ascending=False) sorts the average prices in descending order
# Set the size of the graph
plt.figure(figsize=(16, 15))
# Create the bar graph to show the average price for each county
sns.barplot(x="price",y="County",data=average_price_by_county,palette="viridis")
# Add a title to the graph
plt.title("County Comparison Graph")
plt.xlabel("Average Price")
plt.ylabel("County")
plt.tight_layout()
# this adjusts the layout of the graph to prevent overlapping elements 
# Display the graph
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = pd.read_csv("Cleaned_Data.csv")


# 2. Start with a fresh, large canvas
plt.figure(figsize=(14, 10))

# 3. Create the plot (ignoring empty rows with .dropna())
sns.histplot(data["price"] ,bins=30, color="blue")

# 4. Labels
plt.title("Distribution of Prices in the Dataset", fontsize=14)
plt.xlabel("Price")
plt.ylabel("Frequency")

# 5. Show
plt.show()
```

```python
# The histogram shows that most prices in the data are clustered around the lower end of the price range, with a long tail extending towards higher prices.
# This suggests that while most commodities are priced relatuvely low, there are some outliers with significantly higher prices
```

```python
# Creating a heatmap to show the correlation between price and other numerical variables in the dataset using seaborn
# Select only the numerical columns from the data
numerical_data = data.select_dtypes(include=["float64", "int64"])
# Calculate the correlation matrix
correlation_matrix= numerical_data.corr()
# Create a heatmap to visualize the correlation matrix
# give size to the graph
plt.figure(figsize=(16,10))
sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm", linewidths=0.5)
# Display the graph
plt.show()
```
