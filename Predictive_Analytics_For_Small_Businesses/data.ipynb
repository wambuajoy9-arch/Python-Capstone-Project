```python
# Load the Kenya Food Price dataset from the CSV file 
data = pd.read_csv(f"Kenya_Food_Price.csv")

# Display the first few riws to verify taht the data loaded correctly
print(data.head())
```

```python
# Print the column names to identify the variables in the dataset
print(data.columns)
```

```python
# Getting the general information about the dataset
print(data.info())
```

```python
""" Deleting redundant columns
Renaming the columns 
Dropping duplicates
Cleaning individual columns
Remove the NoN values from the dataset
Check for some more Transformations"""
```

```python
# Filtering the dataset to include only the relevant columns for our analysis 
columns_to_keep = ["date", "County", "price", "commodity", "unit", "market", "category", "latitude", "longitude"]
columns_to_drop = ["market_id", "commodity_id", "currency", "priceflag", "pricetype", "usdprice","County"]
filtered_data = data[columns_to_keep]
print(filtered_data)
print(filtered_data)
```

```python
print(filtered_data.head())
```

```python
# Drop the columns that are not needed for our analysis 
data.drop(columns=columns_to_drop, inplace=True, errors='ignore')
```

```python
data.shape
```

```python
# Renaming the columns to more descriptive names
data.rename(columns={"admin2":"County"}, inplace=True)
```

```python
print(data)
```

```python
new_column_names = []
for i in data.columns:
    new_column_names.append(i.upper())

print(new_column_names)
```

```python
 Creating a copy of the original dataset to keep the original data into account in case we need to refer to it later on 
# This allows modifications of the new dataframe without affecting or loosing the original data
new_dataframe_to_modify = data.copy()
# Renaming the columns to uppercase for better reading and consistenty in the new dataframe
new_dataframe_to_modify.columns = new_column_names
print(new_dataframe_to_modify)#
```

```python
# Print the modified dataframe to verify the changes 
print(new_dataframe_to_modify.head())
```

```python
# Checking for duplicates in the dataset 
new_dataframe_to_modify.duplicated().sum()
```

```python
new_dataframe_to_modify.drop_duplicates()
```

```python
# Remove the NoN values from the dataset
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
county_positions = [i for i, col in enumerate(data.columns) if col == "County"]
for pos in county_positions:
	data.iloc[:, pos] = data.iloc[:, pos].astype(str).str.upper()
# This converts all County columns to uppercase
```
